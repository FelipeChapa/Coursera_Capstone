# Segmenting and Clustering Neighborhoods in Toronto, Canada

## 1. Obtain data and clean data set

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium # map rendering library
import matplotlib.cm as cm #Color map for cluster colors
import matplotlib.colors as colors
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(link)[0]
df
```

```python
#Remove rows with "Not assigned" Borough values
df=df[df['Borough']!="Not assigned"].reset_index(drop=True)
df
```

```python
# All Boroughs have an assigned Neighbourhoods
df[df['Neighbourhood']=="Not assigned"]
```

```python
df.shape
```

## 2. Location data, Latitudes and Longitudes

```python
postal_codes = pd.read_csv('http://cocl.us/Geospatial_data')
postal_codes
```

```python
#Add Latitude and Longitude to original df
df_merged = df.merge(postal_codes)
df_merged
```

## 3. Cluster neighborhoods 
 ### Working with boroughs that contain the word "Toronto". Cluster by Latitude and Longitude

```python
df_toronto = df_merged[df_merged["Borough"].str.contains("Toronto")]
df_toronto
```

```python
kclusters = 5

df_cluster = df_toronto.drop(['Postal Code',"Borough","Neighbourhood"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_cluster)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
df_cluster.head()
```

```python
#Add Clustering labels
df_toronto["Cluster"] = kmeans.labels_
df_toronto.head()
```

```python
#Create map to visualize clusters
map_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighbourhood'], df_toronto['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
